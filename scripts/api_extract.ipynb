# COUNTRIES API with lon and lat

```python

def get_country_latlon():
    url ='https://restcountries.com/v3.1/region/europe?fields=name,latlng'

    try:
        response = requests.get(url)

        if response.status_code == 200:
            posts = response.json()
            return posts
        else:
            print('Error:', response.status_code)
            return None
    except requests.exceptions.RequestException as e:
        print('Error:', e)
        return None

posts = get_country_latlon()

if posts:
    response_country_json = posts
else:
    print('Failed to fetch posts from API.')
```

```python

# creating a dictionnary of european countries and their lat and lon
data_country = response_country_json
dict_country = {}
for country in data_country:
    #accessing the api fields
    country_name = country['name']['common']
    country_latlon = country['latlng']

    dict_country[country_name] = country_latlon
```

```python
for country, latlon in dict_country.items():
    print(country,latlon)
```

# OPEN WEATHER API

```python
import requests
import os
from dotenv import load_dotenv

# Charger les variables d'environnement du fichier .env
load_dotenv()

# Récupérer la clé API
api_key = os.getenv("API_KEY")

def get_country_weather(lat,lon,api_key_):
    url = f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key_}'

    try:
        response = requests.get(url)

        if response.status_code == 200:
            posts = response.json()
            return posts
        else:
            print('Error:', response.status_code)
            return None
    except requests.exceptions.RequestException as e:
        print('Error:', e)
        return None

Paris = [48.864716,2.349014]
posts = get_country_weather(Paris[0],Paris[1],api_key)

if posts:
    response_weather_json = posts
else:
    print('Failed to fetch posts from API.')

print(response_weather_json)
```

```python
def kelvin_to_celsius(kelvin):
    """Converts kalven temperature to celcius and rounding to closest integer"""
    celsius = kelvin - 273.15
    return round(celsius)
```

```python

# Assuming response_json is already a dictionary
data_weather = response_weather_json

#accessing the api fields
coord = data_weather['coord']
lon = coord['lon']
lat = coord['lat']

print(f"City: {data_weather['name']}")
print(f"Coordinates: {lat}, {lon}")
print(f"Weather: {data_weather['weather'][0]['main']} - {data_weather['weather'][0]['description']}")
print(f"Temperature: {kelvin_to_celsius(data_weather['main']['temp'])}°C")
print(f"Humidity: {data_weather['main']['humidity']}%")
print(f"Visibility: {data_weather['visibility']} meters")
print(f"Wind: {data_weather['wind']['speed']} m/s at {data_weather['wind']['deg']}°")
print(f"Sunrise: {data_weather['sys']['sunrise']}, Sunset: {data_weather['sys']['sunset']}")
```

```python
def get_europe_weather_per_country(dict_countries,api_key_):
    dict_weather_europe_per_country_return = {}
    for country,latlon in dict_countries.items():
        lat = latlon[0]
        lon = latlon[1]
        dict_weather_europe_per_country_return[country] = get_country_weather(lat,lon,api_key_)
    return dict_weather_europe_per_country_return

dict_weather_europe_per_country = get_europe_weather_per_country(dict_country,api_key)
```
